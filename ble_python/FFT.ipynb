```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import pi
from scipy.fftpack import fft
```

```python
from numpy import loadtxt
```

```python
times = loadtxt("accdata2.txt", usecols = (0))
pitch = loadtxt("accdata2.txt", usecols = (1))
roll = loadtxt("accdata2.txt", usecols = (2))
pitchLPF = loadtxt("accdata2.txt", usecols = (3))
rollLPF = loadtxt("accdata2.txt", usecols = (4))
len(times)
```

```python
sample_rate = len(times)
N = (2 - 0) * sample_rate
N

times = times#[0:228]
pitch = pitch#[0:228]
roll = roll#[0:228]
pitchLPF = pitchLPF#[0:228]
rollLPF = rollLPF#[0:228]
```

```python
plt.plot (times, pitch)
plt.title ('Pitch Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
plt.plot (times, pitchLPF)
plt.title ('LPF Pitch Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
```

```python
plt.plot (times, roll)
plt.title ('Roll Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
plt.plot (times, rollLPF)
plt.title ('LPF Roll Time Domain Signal')
plt.xlabel ('Time')
plt.ylabel ('Amplitude')
plt.show ()
```

```python
frequency = np.linspace (0.0, sample_rate/2, int (N/2))

freq_data = fft(pitch)
y = 2/N * np.abs (freq_data [0:int (N/2)])
freq_data = fft(pitchLPF)
z = 2/N * np.abs (freq_data [0:int (N/2)])
```

```python
plt.plot(frequency, y)
plt.title('Pitch Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
plt.plot(frequency, z)
plt.title('LPF Pitch Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
```

```python
freq_data = fft(roll)
y = 2/N * np.abs (freq_data [0:int (N/2)])

plt.plot(frequency, y)
plt.title('Roll Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')

plt.show()

freq_data = fft(rollLPF)
y = 2/N * np.abs (freq_data [0:int (N/2)])

plt.plot(frequency, y)
plt.title('LPF Roll Frequency domain Signal')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplitude')
plt.show()
```
